```python

from sklearn.tree import DecisionTreeClassifier
import numpy as np

hour_attendence = np.array([[1,40],[2, 50],[3,60],[4,70],[5,80],[6,90],[7,95]])
fail_pass = np.array([0,0,0,1,1,1,1])

model = DecisionTreeClassifier(random_state=56)
model.fit(hour_attendence,fail_pass)

new_stud = [[7,75]]
prediction = model.predict(new_stud)
print(f"Prediction for the new student is: {'Pass' if prediction[0] == 1 else 'Fail'}")
```
